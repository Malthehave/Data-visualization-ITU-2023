```python
import pandas as pd

# Read the data from xlsx file
df = pd.read_excel('data/accidents.xlsx', sheet_name='93-2014 FINAL')
df.head(20)
```

```python
#Drop columns without "Airbus" or "Boeing" in the plane_type column, capitalization doesn't matter
df = df[df['plane_type'].str.contains("Airbus|Boeing", case=False)]

#change all names containing "Boeing" to "Boeing", and all names containing "Airbus" to "Airbus"
df['plane_type'] = df['plane_type'].str.replace(".*Boeing.*", "Boeing", case=False)
df['plane_type'] = df['plane_type'].str.replace(".*Airbus.*", "Airbus", case=False)

df=df[['date','plane_type','ref','airline','country']]
df.head(20)
```

```python
#save to csv as "accidents_processed.csv"
df.to_csv('data/accidents_processed.csv', index=False)
```
