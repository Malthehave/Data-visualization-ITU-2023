```python
import pandas as pd
```

```python
df = pd.read_csv('data/airline_fleet.csv')
```

```python
df.head()
```

```python
#delete rows with missing values in Current column
df = df.dropna(subset=['Current'])

#delete rows without Airbus or Boeing in "Aircraft Type" column
df = df[df['Aircraft Type'].str.contains("Airbus|Boeing", case=False)]

#map aircraft type containing Airbus to Airbus and Boeing to Boeing
df['Aircraft Type'] = df['Aircraft Type'].map(lambda x: 'Airbus' if 'Airbus' in x else 'Boeing')

#delete "Parent Airline, "Historic", "Orders", "Future", "Total"" columns
df = df.drop(columns=['Parent Airline', 'Historic', 'Orders', 'Future', 'Total'])
```

```python
#rename "Current" column to "Fleet"
df = df.rename(columns={'Current': 'Fleet'})

#change fleet column to integer
df['Fleet'] = df['Fleet'].astype(int)

df.head(1000)
```

```python
df.to_csv('data/airline_fleet_clean.csv', index=False)
```
