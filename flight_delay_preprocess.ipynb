```python
import pandas as pd
```

```python
df = pd.read_csv('data/flights.csv')
```

```python
tail_numbers = df['TAIL_NUMBER']
tail_numbers = tail_numbers.drop_duplicates()
print(tail_numbers.shape)
```

```python
master = pd.read_csv('ReleasableAircraft/MASTER.txt')
master.head()
```

```python
aircraft = pd.read_csv('ReleasableAircraft/ACFTREF.txt')
aircraft.head()
```

```python
aircraft_manufacturer = []
aircraft_model = []
for tail_number in tail_numbers:    
    # Lookup tail number in master
    try:
        row = master[master['N-NUMBER'] == tail_number[1:]]
        # Get manufacturer model code
        code = row['MFR MDL CODE'].values[0]
        # Lookup manufacturer model code in aircraft
        row = aircraft[aircraft['CODE'] == code]
        # Get manufacturer
        manufacturer = row['MFR'].values[0]
        # Get model
        model = row['MODEL'].values[0]        
        aircraft_manufacturer.append(manufacturer.strip())
        aircraft_model.append(model.strip())
    
    except:
        aircraft_manufacturer.append(None)
        aircraft_model.append(None)        
```

```python
# Create new temp df with aircraft_manufacturer, aircraft_model, and tail_number
temp = pd.DataFrame({'aircraft_manufacturer': aircraft_manufacturer, 'aircraft_model': aircraft_model, 'TAIL_NUMBER': tail_numbers})
# Append aircraft_manufacturer and aircraft_model to df based on temp['tail_number']
df = df.merge(temp, on='TAIL_NUMBER', how='left')
df.head()
```

```python
# Remove rows with null values for aircraft_manufacturer
df = df[df['aircraft_manufacturer'].notnull()]
df.shape
```

```python
#join year month and day
df['YEAR_MONTH_DAY'] = df['YEAR'].astype(str) + '-' + df['MONTH'].astype(str) + '-' + df['DAY'].astype(str)
df['YEAR_MONTH_DAY'] = pd.to_datetime(df['YEAR_MONTH_DAY'], format='%Y-%m-%d')
```

```python
df = df[['YEAR_MONTH_DAY','aircraft_manufacturer', 'ARRIVAL_DELAY']] #departure delay doesn't matter as long as the arrival is on time
df.head(20)
```

```python
#delete rows without regex match "AIRBUS" or "BOEING"
df = df[df['aircraft_manufacturer'].str.contains("AIRBUS|BOEING")]

#turn all "AIRBUS INDUSTRIE" to "AIRBUS"
df['aircraft_manufacturer'] = df['aircraft_manufacturer'].str.replace('AIRBUS INDUSTRIE', 'AIRBUS')
df.head(20)
```

```python
#delete rows with 0 "ARRIVAL_DELAY"
df = df[df['ARRIVAL_DELAY'] != 0]

#save all rows with positive "ARRIVAL_DELAY" to "delayed_df"
delayed_df = df[df['ARRIVAL_DELAY'] > 0]

#save all rows with negative "ARRIVAL_DELAY" to "early_df"
early_df = df[df['ARRIVAL_DELAY'] < 0]

#switch early_df to positive
early_df['ARRIVAL_EARLY'] = early_df['ARRIVAL_DELAY'].abs()
early_df = early_df.drop('ARRIVAL_DELAY', axis=1)

#add running sum to both
delayed_df['RUNNING_SUM'] = delayed_df.groupby(['YEAR_MONTH_DAY','aircraft_manufacturer'])['ARRIVAL_DELAY'].cumsum()
early_df['RUNNING_SUM'] = early_df.groupby(['YEAR_MONTH_DAY','aircraft_manufacturer'])['ARRIVAL_EARLY'].cumsum()

delayed_df.head(), early_df.head()
```

```python
#print shapes
print(delayed_df.shape, early_df.shape)
```

```python
#save new dfs to csv
delayed_df.to_csv('data/delayed_flights.csv', index=False)
early_df.to_csv('data/early_flights.csv', index=False)
```
